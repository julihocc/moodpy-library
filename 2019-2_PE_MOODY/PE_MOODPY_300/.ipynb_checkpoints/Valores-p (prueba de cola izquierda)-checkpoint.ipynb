```python
import toMoodle as me
import random as rd
import numpy as np
import sympy as spy
import datetime as dt
import base64
import matplotlib.pyplot as plt
from scipy import stats
#import sys
```

```python
def pruebaDeHipotesis(lista=[], betas = [.01, 0.05, .10]): 
    beta = np.random.choice(betas)
    #print(beta)
    m0 = np.random.poisson(100)
    s = np.random.poisson(10)
    n = np.random.poisson(40)
    x = np.random.normal(m0, s, n)
    m = np.mean(x)
    z = (m-m0)/(s/np.sqrt(n))
    def F(x):
        return stats.norm(0,1).cdf(x)
    valorp = F(z)
    if valorp < beta:
        myMC = "{1:MC:acepta~=rechaza}"
    else:
        myMC = "{1:MC:=acepta~rechaza}"
    
    linea=[]
        
    answers = {"D0":m0, "s":s, "n":n, "D":m, "beta":beta}
    
    ejercicio = r"""
   <p> Considera  la variable aleatoria $$\bar{{X}}=\dfrac{{1}}{{n}}\sum_{{i=0}}^{{n}} X_i$$ 
    donde \(X_i, i=1,...,n\) son variables aleatorias identicamente distribuidas tales que 
    \(E(X_i)=\mu\) y \(Var(X_i)=\sigma^2\). </p>
    
 Dada la siguiente información:
 <ul>
<li>Hipótesis nula \(H_o:\mu=\){a[D0]}</li>
<li>Hipótesis alternativa \(H_a:\mu<\){a[D0]}</li>
<li> <strong> Observe que la desigualdad en la hipótesis alternativa está invertida. </strong></li>
<li>Desviación estándar: \(\sigma=\){a[s]}</li>
<li>Tamaño de la muestra: \(n=\){a[n]}</li>
<li>Valor del estimador: \(\bar{{x}}=\){a[D]}</li>
<li>Nivel de significación: \(\alpha=\){a[beta]}</li>
</ul>
encuentre los siguientes estadísticos: 
    """.format(a = answers)
    
    
    linea.append(ejercicio)
    MC = "<pre>Valor Z="+me.NM(z)+"</pre>"
    MC = MC+"<pre>valor-p="+me.NM(valorp)+"</pre>"
    MC = MC+"<pre>La hipótesis nula Ho se \n"+myMC+"</pre>"

    #print MC
    linea.append(MC)

    #print linea
    lista.append(linea)
```

```python
bloques= 50
ejercicios=[]

for k in range(bloques):
    pruebaDeHipotesis(ejercicios)
mil_exes=[]; mil_sol=[]
for elemento in ejercicios:
    #print elemento
    mil_exes.append(elemento[0])
    mil_sol.append(elemento[1])
me.ExeToMoodle("Valores-p (prueba de cola izquierda)", mil_exes, mil_sol)
```
